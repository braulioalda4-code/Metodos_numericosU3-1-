```python
import numpy as np # Importa la biblioteca NumPy para cálculos numéricos

A = np.array([[10,1,-1], # Matríz de coeficientes del sistema de ecuaciones
              [2,10,1], # Segunda fila de coeficientes
              [2,3,10]]) # Tercera fila de coeficientes

b = np.array([11,12,13]) # Vector de términos independientes del sistema

x = np.linalg.solve(A, b) # Función de NumPy que resuelve el sistema lineal Ax=b

print("Solución del sistema:", x) # Imprime el vector solución del sistema
```
