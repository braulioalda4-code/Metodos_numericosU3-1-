Ejercicio - Método de Gauss-Sidel

```python
import numpy as np # Biblioteca para cáculos numéricos

x = np.zeros (3) # Vector inicial de aproximación para las variables

tol = 1e-6 # Tolerancia permitida para el error

for k in range(100): # Bucle de iteraciones
    x_old = x.copy() # Guarda la solución anterior para calcular el error

    x[0] = (11 - x[1] + x[2]) / 10 # Actualización de la variable x usando valores más reciente

    x[1] = (12 - 2*x[0] - x[2]) / 10 # Actualización de y usando el nuevo valor de x

    x[2] = (13 - 2*x[0] - 3*x[1]) / 10 # Actualización de z usando valores

    if np.linalg.norm(x - x_old) < tol:
        break

print("Solución Gauss-Seidel:", x)
print("Iteraciones:", k+1)
```
