Ejercicio 3 - Aplicación: regresión lineales

```python
import numpy as np # Importa la biblioteca NumPy

X = np.array([[1,1], # Matríz de datos donde la primera columna es el término independiente
              [1,2], # Segunda observación del conjunto de datos
              [1,3], # Tercera observación
              [1,4]]) # Cuarta observación

y = np.array([6,5,7,10]) # Vector de valores observados

XT = X.T # Cálcula la transpuesta de la matríz de datos

beta = np.linalg.inv(XT.dot(X)).dot(XT).dot(y) # Cálcula los coeficientes de la regresión lineal

print("Coeficientes del modelo:", beta) # Imprime los parámetros estimados del modelo
```
