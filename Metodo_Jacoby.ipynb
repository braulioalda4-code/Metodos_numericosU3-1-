Ejercicio 2 - Método de Jacobi

```python
import numpy as np

A = np.array([[10,1,-1],   # Matriz de coeficientes del sistema
              [2,10,1],    # Segunda ecuación
              [2,3,10]])   # Tercera ecuación

b = np.array([11,12,13])   # Vector de términos independientes

x = np.zeros(3)   # Vector inicial de aproximación (x0,y0,z0)

tol = 1e-6   # Tolerancia para el criterio de paro

max_iter = 100   # Número máximo de iteraciones permitidas

for k in range(max_iter):   # Bucle principal de iteraciones

    x_new = np.zeros(3)   # Vector donde se guardará la nueva aproximación

    x_new[0] = (11 - x[1] + x[2]) / 10   # Despeje de x a partir de la primera ecuación

    x_new[1] = (12 - 2*x[0] - x[2]) / 10   # Despeje de y a partir de la segunda ecuación

    x_new[2] = (13 - 2*x[0] - 3*x[1]) / 10   # Despeje de z a partir de la tercera ecuación

    if np.linalg.norm(x_new - x) < tol:   # Verifica si el cambio entre iteraciones es menor que tol
        break   # Termina el proceso si se cumple el criterio de convergencia

    x = x_new   # Actualiza el vector x para la siguiente iteración

print("Solución aproximada:", x)
print("Número de iteraciones:", k + 1)
```
